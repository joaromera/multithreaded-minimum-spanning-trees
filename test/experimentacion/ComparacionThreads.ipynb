```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fb = pd.read_csv("../../file.csv") # resultados nuestros en la macbook air
print(fb)
```

```python
d = {'n': [100], 'grafo': ["arbol"], 'threads': [1], 'tiempo': [1.0]}
data = pd.DataFrame(data=d)
data = data.drop([0])
for k in {1, 2, 4, 8, 16, 32}:
    for j in range(100,1001,100):
        data1 = (((fb.loc[fb['n'] == j]).loc[fb['grafo'] == 'arbol']).loc[fb['threads'] == k]).loc[:, ' tiempo']
        data2 = {'n': [j], 'grafo': ["arbol"], 'threads': [k], 'tiempo': [data1.mean()]}
        data = data.append(pd.DataFrame(data=data2))

        
for k in {1, 2, 4, 8, 16, 32}:
    for j in range(100,1001,100):
        data1 = (((fb.loc[fb['n'] == j]).loc[fb['grafo'] == 'ralo']).loc[fb['threads'] == k]).loc[:, ' tiempo']
        data2 = {'n': [j], 'grafo': ["ralo"], 'threads': [k], 'tiempo': [data1.mean()]}
        data = data.append(pd.DataFrame(data=data2))

for k in {1, 2, 4, 8, 16, 32}:
    for j in range(100,1001,100):
        data1 = (((fb.loc[fb['n'] == j]).loc[fb['grafo'] == 'completo']).loc[fb['threads'] == k]).loc[:, ' tiempo']
        data2 = {'n': [j], 'grafo': ["completo"], 'threads': [k], 'tiempo': [data1.mean()]}
        data = data.append(pd.DataFrame(data=data2))
print(data)
```

```python
#Grafico para comparacion con arboles

data2 = data.loc[data['threads'] == 1]
data3 = data2.loc[data2['grafo'] == 'arbol']
ax1 = sns.lineplot(x="n", y="tiempo", data=data3);

data2 = data.loc[data['threads'] == 2]
data3 = data2.loc[data2['grafo'] == 'arbol']
ax1 = sns.lineplot(x="n", y="tiempo", data=data3);

data2 = data.loc[data['threads'] == 4]
data3 = data2.loc[data2['grafo'] == 'arbol']
ax1 = sns.lineplot(x="n", y="tiempo", data=data3);

data2 = data.loc[data['threads'] == 8]
data3 = data2.loc[data2['grafo'] == 'arbol']
ax1 = sns.lineplot(x="n", y="tiempo", data=data3);

data2 = data.loc[data['threads'] == 16]
data3 = data2.loc[data2['grafo'] == 'arbol']
ax1 = sns.lineplot(x="n", y="tiempo", data=data3);

data2 = data.loc[data['threads'] == 32]
data3 = data2.loc[data2['grafo'] == 'arbol']
ax1 = sns.lineplot(x="n", y="tiempo", data=data3);

plt.ylabel("tiempo(ms)");
plt.xlabel("n");
ax1.legend(["S1", "P2", "P4", "P8", "P16", "P32"]);
```
